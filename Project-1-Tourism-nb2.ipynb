```python
# Krystal's work

#import dependencies + setup
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import datetime as dt 


#study data files 
interst_data= "Data/ye.csv"

#read data
interstate_data= pd.read_csv(interst_data)

#display table for preview 
interstate_data
```

```python
#cleanup
interstate_data.columns=[
    "Year",
    "NSW Arrivals",
    "VIC Arrivals",
    "QLD Arrivals",
    "SA Arrivals",
    "WA Arrivals",
    "TAS Arrivals",
    "NT Arrivals",
    "ACT Arrivals",
    "Other Territories Arrivals",
    "Total Arrivals into AUS"]
interstate_data

intercleaned= interstate_data.drop(range(0,9))
intercleaned

data_index=intercleaned.set_index("Year")
data_index
```

```python
datalast7years= data_index.tail(75)
datalast7years
```

```python
datalast7years['NSW Arrivals'] = datalast7years['NSW Arrivals'].astype(int)
datalast7years['VIC Arrivals'] = datalast7years['VIC Arrivals'].astype(int)
datalast7years['QLD Arrivals'] = datalast7years['QLD Arrivals'].astype(int)
datalast7years['SA Arrivals'] = datalast7years['SA Arrivals'].astype(int)
datalast7years['WA Arrivals'] = datalast7years['WA Arrivals'].astype(int)
datalast7years['TAS Arrivals'] = datalast7years['TAS Arrivals'].astype(int)
datalast7years['NT Arrivals'] = datalast7years['NT Arrivals'].astype(int)
datalast7years['ACT Arrivals'] = datalast7years['ACT Arrivals'].astype(int)
datalast7years['Other Terriroties Arrivals'] = datalast7years['Other Territories Arrivals'].astype(int)
datalast7years['Total Arrivals into AUS'] = datalast7years['Total Arrivals into AUS'].astype(int)
```

```python
#month plot by state
statebymonth= datalast7years.drop(['Total Arrivals into AUS'], axis=1)
lockdowneffects= statebymonth.tail(28)
lockdowneffects
```

```python
nsw=lockdowneffects["NSW Arrivals"]
vic=lockdowneffects["VIC Arrivals"]
qld= lockdowneffects["QLD Arrivals"]
sa= lockdowneffects["SA Arrivals"]
wa= lockdowneffects["WA Arrivals"]
tas= lockdowneffects["TAS Arrivals"]
nt= lockdowneffects["NT Arrivals"]
act= lockdowneffects["ACT Arrivals"]
other = lockdowneffects["Other Terriroties Arrivals"]

plt.style.use('fivethirtyeight')
plt.plot(nsw,label="NSW")
plt.plot(vic, label="VIC")
plt.plot(qld,label="QLD")
plt.plot(sa, label="SA")
plt.plot(wa,label="WA")
plt.plot(tas,label="TAS")
plt.plot(nt, label="NT")
plt.plot(act,label="ACT")
plt.plot(other,label="Other")

plt.legend(loc="best",fontsize=11)


plt.title("Arrivals into Australia by State/Territory by year and month")
plt.xlabel("Year/Month", fontsize= 10)
np.arange(0, 8, 8)
plt.ylabel("Number of Arrivals (n)", fontsize=10)
plt.xticks(np.arange(0, 30, 5),fontsize=12)
plt.yticks(fontsize=12)
plt.style.use('fivethirtyeight')

plt.savefig('Stateandmonth.png')
```

```python
maximum= Interstatebyyear.max()
minimum= Interstatebyyear.min()
decrease= (Interstatebyyear.max()) - (Interstatebyyear.min())
change= round(decrease/(Interstatebyyear.max()) * -100,2)
formatted= change.map("{:,.2f}".format)
formatted

diff = pd.DataFrame({"Max Arrivals":maximum, "Min Arrivals" : minimum, "Percentage of Change (%)" : formatted})
titles=diff.style.set_caption("Maximum and Minimum number of arrivals into Australia per state and percentage of change")
titles
```

```python
plotdif= change.plot.bar()
plt.ylim(-90,-100)
plt.xlabel("State/Terriroty", fontsize=13)
plt.xticks(fontsize=14)
plt.ylabel("Percentage of Change (%)", fontsize=13)
plt.yticks(fontsize=14)
plt.title("Percentage of Change over time in number of arrivals into Australia by State/Territory")
plotdif
plt.style.use('fivethirtyeight')
plt.savefig('percofchange.png')
```
