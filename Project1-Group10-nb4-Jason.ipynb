```python
# Jason's work
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import os
```

```python
# Calculations of how well each Duration category has returned to 'Normal'
avg = []
final = []
ratios =  []

# Find average number of short term visitors for each Duration of Stay for 2016-2019
avg.append(recent_df["Short-term Visitors arriving - Under 1 week"][:48].mean())
avg.append(recent_df["Short-term Visitors arriving - Over 1 and under 2 weeks"][:48].mean())
avg.append(recent_df["Short-term Visitors arriving - Over 2 weeks and under 1 month"][:48].mean())
avg.append(recent_df["Short-term Visitors arriving - Over 1 and under 2 months"][:48].mean())
avg.append(recent_df["Short-term Visitors arriving - Over 2 and under 3 months"][:48].mean())
avg.append(recent_df["Short-term Visitors arriving - Over 3 and under 6 months"][:48].mean())
avg.append(recent_df["Short-term Visitors arriving - Over 6 and under 12 months"][:48].mean())

# Get most recent (March 2022) data for number of visitors for each Duration of Stay
final.append(recent_df["Short-term Visitors arriving - Under 1 week"].iloc[-1])
final.append(recent_df["Short-term Visitors arriving - Over 1 and under 2 weeks"].iloc[-1])
final.append(recent_df["Short-term Visitors arriving - Over 2 weeks and under 1 month"].iloc[-1])
final.append(recent_df["Short-term Visitors arriving - Over 1 and under 2 months"].iloc[-1])
final.append(recent_df["Short-term Visitors arriving - Over 2 and under 3 months"].iloc[-1])
final.append(recent_df["Short-term Visitors arriving - Over 3 and under 6 months"].iloc[-1])
final.append(recent_df["Short-term Visitors arriving - Over 6 and under 12 months"].iloc[-1])

# Calculate the percentage return to normal by doing final/average for each Duration
for i in range(len(avg)):
    ratios.append(100*final[i]/avg[i])
    
# Bar plit data
stays = ["<1 Week", "1-2 Weeks", "2 Weeks - 1 Month", "1-2 Months", "2-3 Months", "3-6 Months", "6-12 Months"]
x = np.arange(len(stays)) # the label locations

width = 0.7 # the width of the bars

plt.ylabel("Percentage return to 'Normal'")
plt.xlabel("Duration of Stay")
plt.title("Percentage Return to 'Normal' by Duration of Stay")
plt.xticks(x, rotation=30)

plt.bar(stays, ratios, width, label='population')
plt.savefig('pltJason-7.png', bbox_inches='tight')
plt.show()
```
